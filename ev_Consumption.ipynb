```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Información sobre el dataset 
print(electric_vehicles.shape)
print(electric_vehicles.info())

# Nulos
print("\nValores nulos:")
print(electric_vehicles.isnull().sum()[electric_vehicles.isnull().sum() > 0])

# Duplicados
print(f"\nDuplicados: {electric_vehicles.duplicated().sum()}")
electric_vehicles = electric_vehicles.drop_duplicates()
```

Eliminar variable Vehicle_Id ya que no aporta nada significativo al estudio

```python
electric_vehicles = electric_vehicles.drop(columns=['Vehicle_ID'])
```

```python

# Revisar valores negativos en variables que no deberían ser negativas

cols_no_negativas = [
    'Speed_kmh',
    'Acceleration_ms2',
    'Battery_State_%',
    'Battery_Voltage_V',
    'Battery_Temperature_C',
    'Slope_%',
    'Humidity_%',
    'Wind_Speed_ms',
    'Tire_Pressure_psi',
    'Vehicle_Weight_kg',
    'Distance_Travelled_km',
    'Energy_Consumption_kWh'
]

for col in cols_no_negativas:
    if col in electric_vehicles.columns:
        negativos = (
            pd.to_numeric(electric_vehicles[col], errors='coerce') < 0
        ).sum()
        print(f"{col}: {negativos} valores negativos")
```

Calcular la variable de correlación de Pearson para ver si existe multicolinealidad entre las variables 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Variables numéricas
df_num = electric_vehicles.select_dtypes(include=np.number)

# Matriz de correlación de Pearson
corr = df_num.corr(method='pearson')

# Mostrar matriz
print(corr)

# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0
)
plt.show()
```

El consumo energético (Energy_Consumption_kWh) está principalmente explicado por la velocidad del vehículo (0.79), que es con diferencia la variable más influyente, seguida por la distancia recorrida (0.47), lo que indica que el modelo depende sobre todo de factores de conducción y uso del vehículo. En un segundo nivel aparecen variables como la humedad (0.17), la temperatura de la batería (0.14), la pendiente (0.09) y la carga ambiental del sistema, que aportan efectos moderados pero secundarios. El resto de variables, incluyendo tráfico, tipo de carretera, modo de conducción y la mayoría de condiciones ambientales, muestran correlaciones cercanas a cero.

Feature engineeriing 

```python
# 1. Diferencia entre temperatura de batería y ambiente
electric_vehicles['Battery_Ambient_Diff'] = (
    electric_vehicles['Battery_Temperature_C'] -
    electric_vehicles['Temperature_C']
)

# 2. Energía consumida por km recorrido
electric_vehicles['Consumption_per_km'] = (
    electric_vehicles['Energy_Consumption_kWh'] /
    electric_vehicles['Distance_Travelled_km']
)


# 4. Índice de resistencia ambiental
electric_vehicles['Environmental_Load'] = (
    electric_vehicles['Wind_Speed_ms'] *
    electric_vehicles['Humidity_%']
)

# 5. Pendiente categorizada
electric_vehicles['Slope_Category'] = pd.cut(
    electric_vehicles['Slope_%'],
    bins=[-100, -3, 3, 100],
    labels=['Downhill', 'Flat', 'Uphill']
)
```

```python
# Variables que quieres correlacionar con consumo energético
features = [
    'Battery_Ambient_Diff',
    'Consumption_per_km',
    'Voltage_per_SOC',
    'Environmental_Load'
]

# Correlación de Pearson con Energy_Consumption_kWh
pearson_corr = electric_vehicles[features + ['Energy_Consumption_kWh']].corr(method='pearson')

# Solo la columna contra Energy_Consumption_kWh
pearson_with_target = pearson_corr['Energy_Consumption_kWh'].sort_values(ascending=False)

print(pearson_with_target)
```

Determinar horas, días de la semana y meses según timestamp 

```python
electric_vehicles['Timestamp'] = pd.to_datetime(electric_vehicles['Timestamp'])

electric_vehicles['Hour'] = electric_vehicles['Timestamp'].dt.hour
electric_vehicles['DayOfWeek'] = electric_vehicles['Timestamp'].dt.dayofweek
electric_vehicles['Month'] = electric_vehicles['Timestamp'].dt.month
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor


# =========================
# VARIABLES
# =========================

num_vars = [
    "Speed_kmh",
    "Acceleration_ms2",
    "Battery_State_%",
    "Battery_Temperature_C",
    "Slope_%",
    "Temperature_C",
    "Humidity_%",
    "Wind_Speed_ms",
    "Tire_Pressure_psi",
    "Vehicle_Weight_kg",
    "Distance_Travelled_km",
    "Battery_Ambient_Diff",
    "Environmental_Load"
]

cat_vars = [
    "Driving_Mode",
    "Road_Type",
    "Traffic_Condition",
    "Weather_Condition",
    "Slope_Category"
]

target = "Energy_Consumption_kWh"


# =========================
# DATASET
# =========================

df_model = electric_vehicles[num_vars + cat_vars + [target]].copy()

X = df_model[num_vars + cat_vars]
y = df_model[target]


# =========================
# PREPROCESSING
# =========================

preprocessor = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_vars),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_vars)
    ]
)


# =========================
# PIPELINE
# =========================

model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=15
)

pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", model)
])


# =========================
# TRAIN / TEST
# =========================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=15
)

pipeline.fit(X_train, y_train)


# =========================
# EVALUACIÓN HOLDOUT
# =========================

y_pred = pipeline.predict(X_test)

print("R²:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))


# =========================
# CROSS VALIDATION
# =========================

scores = cross_val_score(
    pipeline,
    X,
    y,
    cv=5,
    scoring="r2"
)

print(scores)
print("Media R²:", scores.mean())


```

```python

#  GRÁFICO


plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    "r--"
)
plt.xlabel("Real")
plt.ylabel("Predicho")
plt.title("XGBoost con OneHotEncoder")
plt.show()
```

```python
import shap
import numpy as np
import pandas as pd

# =========================
# 1. TRANSFORMAR DATOS
# =========================
X_trans = pipeline.named_steps["preprocessor"].transform(X_test)

# =========================
# 2. MODELO SHAP (XGBOOST)
# =========================
explainer = shap.TreeExplainer(pipeline.named_steps["model"])
shap_values = explainer(X_trans)

# =========================
# 3. IMPORTANCIA
# =========================
importance = np.abs(shap_values.values).mean(axis=0)

# nombres de variables
feature_names = pipeline.named_steps["preprocessor"].get_feature_names_out()

# tabla final
df_importance = pd.DataFrame({
    "feature": feature_names,
    "importance": importance
}).sort_values("importance", ascending=False)

print(df_importance)
```
